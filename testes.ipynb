# Tratamento de dados

```python
df = df[df['Radiacao (KJ/m²)']!=0]
df = df[df['Potência CA (W)']!=0]

df.dropna(subset=['Radiacao (KJ/m²)'], inplace=True)
df.dropna(subset=['Potência CA (W)'], inplace=True) #Excluindo todas as amostras em que a Potência é NaN
```
